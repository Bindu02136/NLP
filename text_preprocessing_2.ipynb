## Bag of Words (BoW)

```python
import pandas as pd
data = pd.read_csv('spam.csv', encoding='latin-1', usecols=[0,1], names = ["label","message"])
data.head()
```

```python
messages = data['message'][:1000]
```

```python
import re
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()

corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]',' ', messages[i])
    review = review.lower()
    review = review.split()
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review= ' '.join(review)
    corpus.append(review)
```

```python
corpus[:10]
```

```python
## Create the Bag OF Words model
from sklearn.feature_extraction.text import CountVectorizer
## for Binary BOW enable binary=True
cv=CountVectorizer(max_features=100,binary=True)
```

```python
X = cv.fit_transform(corpus).toarray()
X.shape, X
```

## N-grams

```python
cv=CountVectorizer(max_features=100,binary=True,ngram_range=(2,3))
X = cv.fit_transform(corpus).toarray()

cv.vocabulary_
```

## TF-IDF

```python
from nltk.stem import WordNetLemmatizer
lemmetizer = WordNetLemmatizer()
```

```python
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]',' ', messages[i])
    review = review.lower()
    review = review.split()
    review=[lemmetizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review= ' '.join(review)
    corpus.append(review)
```

```python
corpus[:10]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=100)
X = tfidf.fit_transform(corpus).toarray()
```
